## variable to file dictionary

```python
var_to_file = {
    'rhumidity':'humid_monthly_admin2.rds',
    'rhumidity_pop_weighted':'humid_monthly_admin2.rds',
    'precipitation_mm':'prec_monthly_admin2.rds',
    'precipitation_mm_pop_weighted':'prec_monthly_admin2.rds',
    'temperature_celsius':'temp_monthly_admin2.rds',
    'temperature_celsius_pop_weighted':'temp_monthly_admin2.rds',
    'cases':'data_cases_deaths_pop_2016_2024_38.rds',
    'deaths':'data_cases_deaths_pop_2016_2024_38.rds',
    'surveillance_pop_pct':['surveillance_admin2.rds', 'urban_surveillance_admin2.rds'],
    'surveillance_area_pct':['surveillance_admin2.rds', 'urban_surveillance_admin2.rds'],
    'urban_pop_pct_binary':'urbanisation_admin2.rds',
    'urban_area_pct_binary':'urbanisation_admin2.rds',
    'ONI':'ONI.csv'
}

# save dictionary to file such as a csv with one key column and one value column
df_var_to_file = pd.DataFrame(list(var_to_file.items()), columns=['variable', 'file_name'])
df_var_to_file.to_csv('var_to_file.csv', index=False)
```

```python
# how do i read the saeved csv back into a dictionary?
df_loaded = pd.read_csv('var_to_file.csv')
var_to_file_loaded = dict(zip(df_loaded['variable'], df_loaded['file_name']))
print(var_to_file_loaded)
```

```python
# make all strings uppercase in both columns
wikipedia_provinces['Indonesian Name'] = wikipedia_provinces['Indonesian Name'].str.upper()
wikipedia_provinces['Capital'] = wikipedia_provinces['Capital'].str.upper()
province_to_capital = dict(zip(wikipedia_provinces['Indonesian Name'], wikipedia_provinces['Capital']))
# save this to a csv file
df_province_to_capital = pd.DataFrame(list(province_to_capital.items()), columns=['province', 'capital'])
df_province_to_capital.to_csv('province_to_capital.csv', index=False)
```

```python
df_loaded = pd.read_csv('province_to_capital.csv')
province_to_capital_loaded = dict(zip(df_loaded['province'], df_loaded['capital']))
print(province_to_capital_loaded)
```
