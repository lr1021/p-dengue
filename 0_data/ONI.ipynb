```python
import numpy as np
import pandas as pd
import os
```

```python
folder = "../../_data/p-dengue/ONI"
os.listdir(folder)
```

```python
# Read the space-separated text file
df = pd.read_csv(os.path.join(folder, "ONI_clean.txt"), 
                 sep=r"\s+",    # split on any whitespace
                 header=None,              # no header row
                 comment='\\')             # ignore the trailing '\'

# Drop any empty columns (just in case)
df = df.dropna(axis=1, how='all')

# Save as CSV
df.to_csv(os.path.join(folder, "ONI.csv"), index=False, header=False)
```
