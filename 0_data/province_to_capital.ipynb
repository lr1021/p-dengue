```python
# laptop
folder = "../../_data/p-dengue/wikipedia_provinces"
os.listdir(folder)
```

```python
wikipedia_provinces = pd.read_csv(os.path.join(folder,'wikipedia_provinces.csv'))
```

```python
wikipedia_provinces.head()
```

```python
wikipedia_provinces['Indonesian Name'] = wikipedia_provinces['Indonesian Name'].str.upper()
wikipedia_provinces['Capital'] = wikipedia_provinces['Capital'].str.upper()
province_to_capital = dict(zip(wikipedia_provinces['Indonesian Name'], wikipedia_provinces['Capital']))
```

```python
df_province_to_capital = pd.DataFrame(list(province_to_capital.items()), columns=['province', 'capital'])
df_province_to_capital.to_csv(os.path.join(folder, 'province_to_capital.csv'), index=False)
```
